```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the dataset from CSV file
df=pd.read_csv('/content/StudentsPerformance.csv')
df
```

```python
# Select the first 50 rows for analysis
df=df.head(50)
df
```

```python
# Convert the 'math score' column to a NumPy array
maths_score=df['math score'].to_numpy()
maths_score
```

```python
# Convert the 'reading score' column to a NumPy array
reading_score=df['reading score'].to_numpy()
reading_score
```

```python
# Convert the 'writing score' column to a NumPy array
writing_score=df['writing score'].to_numpy()
writing_score
```

```python
# Calculate total marks for each student by adding all three subjects
total=maths_score+reading_score+writing_score
total[:10]
```

```python
# Calculate the average score for each subject
average=[np.mean(maths_score),np.mean(reading_score),np.mean(writing_score)]
average
```

```python
# Average score by gender
gender_average=df.groupby('gender')[['math score','reading score','writing score']].mean()
gender_average
```

```python
# Plot gender-wise average
gender_average.plot(kind='bar',figsize=(6,4),color=['darkorange','grey','seagreen'])
plt.title('Average Scores by Gender')
plt.ylabel('Average Score')
plt.legend(title='Subjects')
plt.show()
```

```python
# Group by parental level of education
parent_average = df.groupby('parental level of education')[['math score','reading score','writing score']].mean()
parent_average
```

```python
# Plot
parent_average.plot(kind='bar', figsize=(10,5), color=['grey','lightblue','steelblue'])
plt.title('Average Scores by Parental Education Level')
plt.ylabel('Average Score')
plt.xlabel('Parental Education')
plt.legend(title='Subjects')
plt.show()
```

```python
# Group by test preparation course
testprep_average=df.groupby('test preparation course')[['math score','reading score','writing score']].mean()
testprep_average
```

```python
# Plot
testprep_average.plot(kind='bar', figsize=(6,4))
plt.title('Average Scores by Test Preparation Course')
plt.ylabel('Average Score')
plt.xlabel('Test Preparation')
plt.legend(title='Subjects')
plt.show()
```

```python
# Group by lunch type
lunch_average=df.groupby('lunch')[['math score','reading score','writing score']].mean()
lunch_average
```

```python
# Plot
lunch_average.plot(kind='bar', figsize=(6,4), color=['blue','green','orange'])
plt.title('Average Scores by Lunch Type')
plt.ylabel('Average Score')
plt.xlabel('Lunch Type')
plt.legend(title='Subjects')
plt.show()
```

```python
# Plot a bar chart of average scores for Math, Reading, and Writing
average=[np.mean(maths_score),np.mean(reading_score),np.mean(writing_score)]
subjects=['Math','Reading','Writing']
plt.bar(subjects,average,color=['blue','green','darkorange'])
plt.title('Average Scores')
plt.ylabel('Average Score')
plt.show()
```
